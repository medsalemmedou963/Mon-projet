```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import gradio as gr
def previsions_meteo_interactives(lieu):
 # Charger les donnes de prvision mtorologique  l'aide de pandas
        df = pd.read_csv("Temperature.csv")  # Remplacez par le chemin rel de votre ensemble de donnes de prvision mtorologique
   
        def afficher_previsions_meteo(lieu):
        # Filtrer les donnes en fonction du lieu saisi par l'utilisateur
            df_lieu = df[df['Lieu'] == lieu]

        # Crer des graphiques interactifs  l'aide de matplotlib
      
            plt.figure(figsize=(10, 6))
            fig, ax = plt.subplots(figsize=(20, 10))
            plt.scatter(x=df_lieu['TMAX'], y=df_lieu['TMIN'])
            plt.plot(df_lieu['DATE'], df_lieu['Lieu'], label='Lieu')   
            plt.plot(df_lieu['DATE'], df_lieu['ELEVATION'], label='ELEVATION')
            plt.plot(df_lieu['DATE'], df_lieu['LATITUDE'], label='LATITUDE')
            plt.plot(df_lieu['DATE'], df_lieu['LONGITUDE'], label='LONGITUDE')
            plt.plot(df_lieu['DATE'], df_lieu['DATE'], label='DATE')
            plt.plot(df_lieu['TMIN'], df_lieu['TMAX'], label='TMAX')
            plt.plot(df_lieu['DATE'], df_lieu['TMIN'], label='TMIN')
            plt.plot(df_lieu['DATE'], df_lieu['PRCP'], label='PRCP')
            plt.xlabel("DATE")
            plt.ylabel("Valeur")
            plt.title(f"Prvisions mtorologiques pour {lieu}")
            plt.legend()
            plt.xticks(rotation=45)
            plt.show()
         

   # Crer l'interface Gradio
        interface = gr.Interface(
            fn=afficher_previsions_meteo,
            inputs=gr.inputs.Textbox(),
            outputs=['plot'], 
            article="Talk more about Penguins here, shall we?",
            theme='peach',
            title="Prvisions Mto Interactives",
            description="Entrez un lieu et visualisez des graphiques ou des diagrammes interactifs, montrant les prvisions de temprature, d'humidit et de prcipitations.",
            examples=[["Paris"]],)
        afficher_previsions_meteo('Paris')

        # Lancer l'interface
        interface.launch()
previsions_meteo_interactives("paris")
```
